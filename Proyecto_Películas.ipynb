```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/GabiMiranda05/data-science-portfolio/refs/heads/main/Ejercicios-python/netflix_titles.csv', encoding = 'latin-1')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Muestro la suma de nulos por columna
df.isnull().sum()
```

```python
# Muestra cantidad de filas y columnas
filas, columnas = df.shape
print(f"El DataFrame tiene {filas} filas y {columnas} columnas.")
```

```python
df.describe()
```

```python
# Eliminar columnas desde Unnamed: 12 a Unnamed: 25
df_copia = df.copy()
df_copia = df_copia.drop(df_copia.columns[12:26], axis=1)
df_copia.head()
```

```python
# Se calcula la moda en la columna rating
moda_rating = df_copia['rating'].mode()[0]
print(moda_rating)
```

```python
# Completar rating con valor promedio
df_copia['rating'].fillna(moda_rating, inplace=True)
```

```python
df_copia['duration'].unique()
```

```python
# Dividir la columna duración por serie y película
df_copia['duration_min'] = df_copia['duration'][df_copia['duration'].str.contains('min', na=False)]
df_copia['duration_season'] = df_copia['duration'][df_copia['duration'].str.contains('Season|Seasons', na=False)]
```

```python
df_copia.head()
```

```python
df_copia['duration_season'] = df_copia['duration'].str.extract('(\d+)\s+(?:Season|Seasons)', expand=False)
df_copia['duration_min'] = df_copia['duration'].str.extract('(\d+)\s+(?:min)', expand=False)
```

```python
df_copia.head()
```

```python
df_copia = df_copia.dropna(subset=['duration'])
```

```python
df_copia.isnull().sum()
```

```python
# Gráfico barras de Cantidad de peliculas adheridas a Netflix VS Cantidad de Series
# Conteo series
Conteo_series = df_copia['type'].value_counts()

sns.barplot = sns.barplot(x=Conteo_series.index, y=Conteo_series.values)
sns.barplot.set_title('Cantidad de peliculas adheridas a Netflix VS Cantidad de Series')
```

```python
df_copia['release_year'].max()
```

```python
df_copia['release_year'].min()
```

```python
df_copia['release_year_group'] = np.where(df_copia['release_year'] < 2012, 'Antes_de_2012', df_copia['release_year'])
```

```python
df_copia['release_year_group'].unique()
```

```python
# Gráfico de barras de Cantidad de peliculas realizadas en función del año
Conteo_years = df_copia['release_year_group'].value_counts()
plt.pie(Conteo_years.values, labels=Conteo_years.index, autopct='%1.1f%%')
plt.title('Cantidad de películas lanzadas por año')
plt.show()
```
